# 산탄데르 데이터 EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 판다스 사이즈 설정

# 최대 줄 수 설정
pd.set_option('display.max_rows', 50)
# 최대 열 수 설정
pd.set_option('display.max_columns', 50)
# 표시할 가로의 길이
pd.set_option('display.width', 100)
```

```python
# 파일 읽어오기

df = pd.read_csv("../input/train_ver2.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# 결측수

df.isnull().sum()
```

```python
# 고객변수 중 수치형 변수 살펴보기

num_cols = [col for col in df.columns[:24] if df[col].dtype in ['int64', 'float64']]
df[num_cols].describe()
```

```python
# 고객변수 중 범주형 변수 살펴보기

cat_cols = [col for col in df.columns[:24] if df[col].dtype in ['O']]
df[cat_cols].describe()
```

```python
# 범주형 변수 고유값 출력해보기

for col in cat_cols:
    uniq = df[col].unique() # .str.strip()//.str은 str타입데이터에 문자열 함수를 적용시킬때 사용
    print('-' * 50)
    print(f'# col: {col}, n_uniq: {len(uniq)}, uniq: {uniq}')
```

```python
df["age"] = pd.to_numeric(df["age"], errors = 'coerce') # 나이변수를 숫자형으로
df["antiguedad"] = pd.to_numeric(df["antiguedad"], errors = 'coerce') # 거래 누적 기간을 숫자형으로
df["indrel_1mes"] = pd.to_numeric(df["indrel_1mes"], errors = 'coerce') # 월초 기준 고객등급을 숫자형으로
```

```python
df.isnull().sum()
```

```python
# 결측치가 많은 변수 삭제

df = df.drop(["ult_fec_cli_1t","conyuemp"], axis = 1)
```

```python
# 지방이름 기준으로 가구 총수입의 중앙값을 구함

grouped = df.groupby("nomprov").agg({"renta":lambda x: x.median(skipna=True)}).reset_index()
grouped
```

```python
# 'df'랑 'grouped'를 지방이름을 기준으로 합침, 이때 renta이름이 같이 존재하여 'renta_x', 'renta_y'로 나눠짐

new_incomes = pd.merge(df,grouped,how="inner",on="nomprov").loc[:, ["nomprov","renta_y"]] # "renta_y"는 renta변수가 같은 이름이기 때문에 _x와 _y로 분리되어 나타남.
new_incomes
```

```python
# 'renta_y'를 'renta'로 이름을 변경

new_incomes = new_incomes.rename(columns={"renta_y":"renta"}).sort_values("renta").sort_values("nomprov")
new_incomes
```

```python
# 지방이름 기준으로 정렬

df.sort_values("nomprov",inplace=True)
df
```

```python
# 인덱스 번호를 새로 정함

df = df.reset_index()
df
```

```python
# 인덱스 번호를 새로 정함

new_incomes = new_incomes.reset_index()
new_incomes
```

```python
# 'df'에 'new_incomes'의 renta값을 대입 isnull()에서 True인것에 만 해당

df.loc[df.renta.isnull(),"renta"] = new_incomes.loc[df.renta.isnull(),"renta"].reset_index()
df
```

```python
# 지방이름이 없는 'renta' 값에도 구해 놨던 'renta'중앙값이 포함된 전체 'renta'에 대한 중앙값을 대입

df.loc[df.renta.isnull(),"renta"] = df.loc[df.renta.notnull(),"renta"].median()
df
```

```python
# 날짜 기준으로 정렬

df.sort_values(by="fecha_dato",inplace=True)
df
```

```python
# 결측있는 행 삭제 # index열 삭제

df = df.dropna(axis=0)
del df['index']
```

```python
df.isnull().sum()
```

```python
# 필요없는 객체 삭제

del new_incomes, grouped, num_cols, col, cat_cols, uniq
```

```python
df.info()
```

```python
# 은행 거래 누적 기간이 -99999인것이 있음. 이것을 0으로 바꿈

df[df["antiguedad"]<0] = 0
```

```python
# 상품 총 개수를 나타내는 변수생성

df["tot_products"] = df.loc[:,"ind_ahor_fin_ult1":"ind_recibo_ult1"].sum(axis=1)
```

```python
# 상품변수 데이터 축소

for col in df.columns[22:]:
    df[col]=df[col].astype('int8')
```

```python
# 위에서도 확인했지만 거주국가의 종류가 113개에서 결측을 제거하니 6개로 줄었음 이 6개 중에서 제일 많은 ES를 확인

df['pais_residencia'].value_counts()
```

```python
# 거주국가가 'ES'인 데이터만 남김

df = df.loc[lambda df: df.pais_residencia == "ES", :]
```

```python
# 이 변수도 'N'이 압도적으로 많은 것을 볼 수 있음

df['ind_empleado'].value_counts()
```
